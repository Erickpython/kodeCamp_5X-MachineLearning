```python
import numpy as np
import time
```

```python
# Define a large array of numbers from 0 to 1,000,000 without using NumPy's vectorized operations

x = np.array(list(range(1000001)))

y = []

start = time.time()
for i in x:
    y.append(i**2)
end = time.time()
print("Time taken for loop: ", end - start)
```

```python
n = 1000000
a = np.random.rand(n)
b = np.random.rand(n)

start = time.time()
c = []
for i in range(n):
    c.append(a[i] + b[i])
end = time.time()
print("Time taken for loop addition: ", end - start)
```

```python
n = 1000000
a = np.random.rand(n)
b = np.random.rand(n)
start = time.time()
c = a - b
end = time.time()
print("Time taken for vectorized subtraction: ", end - start)
```
