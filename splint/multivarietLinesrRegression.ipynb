# MULTIVARIABLE LINEAR REGRESSION 

Y = B + B1X1 + B2X2 + B3X3 + ... + BnXn + E

```python
# linear regression with a soingle variable
# y = mx + c

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array(np.random.rand(100, 1) * 10)
y = np.array(2.5 * x + 1 + np.random.rand(100, 1) * 5)
```

```python
#parameters
m = 0.0
c = 0.0
learning_rate = 0.01
epochs = 100
losses = []
```

```python
#calculate predictions 
def y_pred (x , m , c):
    return m * x + c


#calculate cost function
def cost_function (x , y , m , c):
    return np.mean((y_pred(x , m , c) - y) ** 2)


#calculate gradients and update parameters
def gradient_descent (x , y , m , c , learning_rate):
    n = len(x)
    error = y_pred(x , m , c) - y
    m_gradient = (-2/n) * np.sum((error * x))
    c_gradient = (-2/n) * np.sum(error)
    return m_gradient , c_gradient
```

```python
for i in range(epochs):
    loss = cost_function(x , y , m , c)
    losses.append(loss)
    m_gradient , c_gradient = gradient_descent(x , y , m , c , learning_rate)
    m = m + learning_rate * m_gradient
    c = c + learning_rate * c_gradient
    if i % 10 == 0:
        print(f"Epoch {i} - Cost: {cost_function(x , y , m , c)}")

#plotting the results
plt.figure (figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(losses)
plt.xlabel('Epochs')


plt.subplot(1, 2, 2)
plt.scatter(x , y , color='blue' , label='Data Points')
plt.plot(x , y_pred(x , m , c) , color='red' , label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')

plt.show()
```

```python
# step 1 : initialize the parameters 
m, c = 0.0 , 0.0
learning_rate = 0.01
epochs = 100
losses = []

#step 2: train the model
for i in range(epochs):
    y_pred = m * x + c
    error = y_pred - y
    cost = np.mean(error ** 2)
    dm = -2 * np.mean(error * x)  #dj/dm
    dc = -2 * np.mean(error)       #dj/dc


    m = m + learning_rate * dm
    c = c + learning_rate * dc
    losses.append(cost)
    

#plotting the results
plt.figure (figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(losses)
plt.xlabel('Epochs')


plt.subplot(1, 2, 2)
plt.scatter(x , y , color='blue' , label='Data Points')
plt.plot(x , y_pred , color='red' , label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')

plt.show()
```

```python
x1 = np.array(np.random.rand(100, 1) * 10)
x2 = np.array(np.random.rand(100, 1) * 10)
y = np.array(2.5 * x1 + 1.5 * x2 + 1 + np.random.rand(100, 1) * 5)
```
