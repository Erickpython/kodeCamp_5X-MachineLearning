```python
# linear regression assignment y = mx + c
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([ 0.        ,  0.52631579,  1.05263158,  1.57894737,  2.10526316,
     2.63157895,  3.15789474,  3.68421053,  4.21052632,  4.73684211,
     5.26315789,  5.78947368,  6.31578947,  6.84210526,  7.36842105,
     7.89473684,  8.42105263,  8.94736842,  9.47368421, 10.        ])

Y = np.array([ 2.60943416,  1.76213705,  7.18511292,  9.40744522,  5.46635068,
     8.6061673 , 13.30831239, 14.26225166, 16.70323979, 16.87285951,
    22.17984858, 23.81874177, 24.23732455, 28.20185083, 28.72449237,
    27.91299402, 32.21118578, 31.39802427, 36.91479534, 36.90014818])
```

```python
# parameters 
m = 0
c = 0
lr = 0.01
epochs = 1000
n = len(X)
costs = []
```

```python
# predict 
def predict(m, X, c):
    return m *  X + c

def mse(Y, y_pred):
    return np.mean((Y - y_pred)**2)
```

```python
for i in range(epochs):

    y_pred = predict(m, X, c)
    error = Y - y_pred
    cost = mse(Y, y_pred)
    costs.append(cost)

    dm = (-2/n) * sum(X * (Y - y_pred))
    dc = (-2/n) * sum(Y - y_pred)

    m = m - lr * dm
    c = c - lr * dc

    if (i+1) % 10 == 0:
        print(f"Epoch {i+1}: m={m}, c={c}, mse={mse(Y, y_pred)}")
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(X, Y)
plt.plot(X, predict(m, X, c), color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Fit')

plt.subplot(1, 2, 2)
plt.plot(costs)
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.title('Cost Function over Epochs')  


plt.show()
```
