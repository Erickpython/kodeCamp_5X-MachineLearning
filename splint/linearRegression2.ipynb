# GRADIENT DESCENT 

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# make sample data 
def make_sample_data(n=20, seed=42, slope=3.5, intercept=2.0, noise_std=2.0):
    rng = np.random.default_rng(seed)
    x = np.linspace(0, 10, n)
    noise = rng.normal(loc=0.0, scale=noise_std, size=n)
    y = slope * x + intercept + noise
    return x, y
```

```python
x,y = make_sample_data()
print(x)
print(y)
```

```python
def pred(m, x, c):
    y_pred = m * x + c
    return y_pred
```

```python
def compute_mse(y_true, y_pred):
    mse = np.mean((y_true - y_pred)**2)
    return mse
```

```python
def gradients(x, y, m, c):
    n = x.shape[0]
    y_hat = m * x + c
    error = y - y_hat
    dm = (-2/n) * np.sum(x * error)
    dc = (-2/n) * np.sum(error)
    return dm , dc 
```

```python
lr = 0.01
m = 0 
c = 0 
ms = []
cs = []
errors =[]
```

```python
y_hat = pred(x, m , c)
error = compute_mse (y, y_hat)
grad = gradients(x, y, m , c)
print (f"Error: {error}\n")
print(f"Gradient: {grad}\n")

plt.figure(figsize=(12, 5))
plt.subplot(1,2,1)
plt.plot(x,y, 'o')
plt.plot(x, y_hat, "--", color="red")

plt.subplot(1,2,2)
ms.append(m)
cs.append(c)
errors.append(error)
plt.plot(errors, "o")


# adjust parameters
m = m - lr*grad[0]
c = c - lr*grad[1]
print(f"Updated Parameters: m = {m}\n c = {c}")
```
