```python
#linear regression  y_pred = mx + c
# to create a linear model that predicts y (y_pred) we define constant variables m and c
m, c = 3.5, 2.5
y_=[]

# a function to run the mathematical model
def y_pred(x):
  for i in x:
    y_.append(i * m + c)
  return y_
(y_pred(x))
```

```python
plt.scatter (x,y)
plt.plot(x, y_, color="red")
plt.title('Linear Regression')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
# mean squared error of the model with constant parameters m and c of your choice
def loss_function(y, y_):
  squared_errors = []
  for i in range (len(y)):
    error = y[i] - y_[i]
    squared_errors.append(error ** 2)
  mse = sum(squared_errors) / len(y)
  return mse

mse = loss_function(y, y_)
print(f"Mean Squared Error: {mse}")
```
