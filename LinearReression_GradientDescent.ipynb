```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Collecting data and turning it to numpy array

x = np.array([ 0. , 0.52631579, 1.05263158, 1.57894737, 2.10526316, 2.63157895, 3.15789474,
             3.68421053, 4.21052632, 4.73684211, 5.26315789, 5.78947368, 6.31578947,
              6.84210526, 7.36842105, 7.89473684, 8.42105263, 8.94736842, 9.47368421, 10. ])

y = np.array([ 2.60943416, 1.76213705, 7.18511292, 9.40744522, 5.46635068, 8.6061673 , 13.30831239,
             14.26225166, 16.70323979, 16.87285951, 22.17984858, 23.81874177, 24.23732455,
              28.20185083, 28.72449237, 27.91299402, 32.21118578, 31.39802427, 36.91479534, 36.90014818])
```

```python
# initialize the parameters to be used in training the models
m = 0   # initial gradient
c = 0   # initial y intercept
lr = 0.001  # steps during model training
epochs = 1000   # number of itterarions - learning cycles
n = len(x)
```

```python
# functions for predicting y_ and mse

def predict(x, m, c):
  return x * m + c


def mse(y, y_pred):
  return np.mean((y - y_pred)**2)
```

```python
# gradient computation
cost_history = []

for epoch in range (epochs):
  y_pred = predict(x, m, c)
  error = y - y_pred

  # gradients dj/dm and dj/dc
  dm = (2/n) * np.dot(error, x)
  dc = (2/n) * np.sum(error)

  # update learning parameters
  m = m + lr * dm
  c = c + lr * dc

  # compute mse
  cost = mse(y, y_pred)
  cost_history.append(cost)

  # print mse after each epoch
  print(f"Epoch {epoch+1}: MSE = {cost:.4f}")
```

```python
# display final parameters
print(f"\nFinal Model Parameters:")
print(f"Slope (m) = {m:.4f}")
print(f"Intercept (c) = {c:.4f}")
```

```python
# plotting - cost function v training progress
plt.figure(figsize=(8,5))
plt.plot(range(len(cost_history)), cost_history, "b-")
plt.title("MSE OVER TRAINING ITTERATIONS")
plt.xlabel("Epoch")
plt.ylabel("MSE")
plt.grid(True)
plt.show()
```

```python
# plotting the regression line v actual data
plt.figure(figsize=(10,6))
plt.scatter(x, y, color="red", label="Actual Data")
plt.plot(x, predict(x, m, c), color="blue", label="Regression Line")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.grid(True)
plt.show()
```
